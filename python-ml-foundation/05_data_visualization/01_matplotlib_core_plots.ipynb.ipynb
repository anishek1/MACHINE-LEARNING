## Skills Gained
- Data visualization

```python
x=[1,2,3,4,5,6,7]
y=[11,22,33,44,55,66,77]

##create a line plot
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("basic line plot")
```

```python
x=[1,2,3,4,5,6,7]
y=[11,22,33,44,55,66,77]

##create a customized line plot

plt.plot(x,y,color='green',linestyle='--',marker='o',markersize=6,linewidth=2)
plt.grid(True)
```

```python
#multiple plots
#sample data

x=[1,2,3,4,5,6,7]
y=[11,22,33,44,55]
y1=[23,32,54,65,23]

plt.figure(figsize=(9,5))

# use lowercase subplot and make sure x and y have matching lengths where needed
x5 = x[:len(y)]  # trim x to match length of y (5 elements)

plt.subplot(2,2,1)
plt.plot(x5, y, color='red')
plt.title("PLOT ")

plt.subplot(2,2,2)
plt.plot(x5, y1, color='green')
plt.title("PLOT 2")

plt.subplot(2,2,3)
plt.plot(y1, y, color='yellow')
plt.title("PLOT 3")

plt.subplot(2,2,4)
plt.plot(y1, x5, color='blue')
plt.title("PLOT 4")

plt.tight_layout()
plt.show()
```

```python
#bar plot
categories=['A','b','C']
values=[44,55,55]

plt.bar(categories,values,color='purple')
plt.xlabel("categories")
plt.ylabel("values")
plt.title('bar plot')
plt.show()
```

```python
#histogram

data=[1,1,2,2,3,3,4,34,45,4,5,56,5,6]
#plot a histogram

plt.hist(data,bins=66)
```

```python
#create a scattter plot
#sample data

x=[12,23,34,45,54]
y=[43,45,32,56,76]
plt.scatter(x,y)
```

```python
##pie chart

labels=['a','b','c','d','e']
sizes=[23,43,32,4,3]
colors=['gold','yellow','red','green','blue']
explode=(0.2,0,0.8,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
## Sales Data Visualization

import pandas as pd

sales_data_df=pd.read_csv('data.csv')
sales_data_df.head(5)
```

```python
sales_data_df.info()
```

```python
#plot total sales 

total_sale_by_product=sales_data_df.groupby('Product Category')['Total Revenue'].sum()
print(total_sale_by_product)
```

```python
total_sale_by_product.plot(kind='bar',color="Pink")
```

```python
## plot sales trend over time

sales_trend=sales_data_df.groupby('Date')['Total Revenue'].sum()
print(sales_trend)
```

```python
sales_trend.plot()
```
