```python
import seabor as sns
```

```python
##ploting with seaborn(basic)

tips=sns.load_dataset('tips')
print(tips)
```

```python
## create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
##categorial plots
##bar plot

sns.barplot(x='total_bill',y='smoker',data=tips)
plt.title("Total Bill vs smoker")
plt.show()
```

```python
##box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
#histograms
sns.histplot(x='day',y='total_bill',data=tips,bins=10)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
#pairplot

sns.pairplot(tips)
```
